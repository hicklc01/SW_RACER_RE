```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot(f):
    x = np.linspace(-6, 6, 300)
    y = []
    for e in x:
        y.append(f(e))

    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Plot of the Function')
    plt.grid(True)
    plt.show()

def plot2(f):
    x = np.linspace(-6, 6, 300)
    y = np.linspace(-6, 6, 300)

    X, Y = np.meshgrid(x, y)
    z2 = np.vectorize(f)
    Z = z2(X, Y)
    Z2 = f(X, Y)

    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.contour3D(X, Y, Z, 500)
    ax.contour3D(X, Y, Z2, 500, cmap='binary')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
```
